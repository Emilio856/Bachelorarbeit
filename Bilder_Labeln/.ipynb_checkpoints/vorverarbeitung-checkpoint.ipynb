```python
import tensorflow as tf
import os
import matplotlib.image as mpimg
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# As in https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_features_harris/py_features_harris.html

img = cv2.imread("30grad.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

gray = cv2.equalizeHist(gray)

gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)

dst = cv2.dilate(dst, None)
dst = cv2.erode(dst, None)

img[dst>0.01*dst.max()] = [0,0,255]

#cv2.imshow("dst", img)
#cv2.waitKey()

plt.imshow(img, cmap="gray")
```

```python
# SubPixel Accuracy
img = cv2.imread("40grad.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(gray)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))   # clipLimit = 2.0 or less. 1.0 works also fine
cl1 = clahe.apply(gray)

# find Harris corners
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 5, 3, 0.04)
kernel = np.ones((5, 5), np.uint8)
dst = cv2.morphologyEx(dst, cv2.MORPH_CLOSE, kernel)
dst = cv2.morphologyEx(dst, cv2.MORPH_OPEN, kernel)
ret, dst = cv2.threshold(dst, 0.01*dst.max(), 255, 0)
dst = np.uint8(dst)

# find centroids
ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)
# define the criteria to stop and refine the corners
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 0.01)
corners = cv2.cornerSubPix(gray, np.float32(centroids), (5, 5), (-1, -1), criteria)

# Now draw them
res = np.hstack((centroids, corners))
res = np.int0(res)
#img[res[:,1], res[:,0]] = [0, 0, 255]
img[res[:,3], res[:,2]] = [255, 0, 0]

plt.subplot(121), plt.imshow(gray, cmap="gray")
plt.subplot(122), plt.imshow(dst, cmap="gray")
cv2.imwrite("subpixel5.png", img)
```

```python
"""
Another option
"""
# SubPixel Accuracy

img = cv2.imread("70grad.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

gray = cv2.equalizeHist(gray)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))   # clipLimit = 2.0 or less. 1.0 works also fine
cl1 = clahe.apply(gray)

# find Harris corners
cl1 = np.float32(cl1)
dst = cv2.cornerHarris(cl1, 2, 3, 0.04)
kernel = np.ones((5, 5), np.uint8)
dst = cv2.morphologyEx(dst, cv2.MORPH_CLOSE, kernel)
dst = cv2.morphologyEx(dst, cv2.MORPH_OPEN, kernel)
ret, dst = cv2.threshold(dst, 0.005*dst.max(), 255, 0)
dst = np.uint8(dst)

# find centroids
ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)

# define the criteria to stop and refine the corners
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)
corners = cv2.cornerSubPix(cl1, np.float32(centroids), (5, 5), (-1, -1), criteria)

# Now draw them
res = np.hstack((centroids, corners))
res = np.int0(res)
img[res[:,1], res[:,0]] = [0, 0, 255]
img[res[:,3], res[:,2]] = [0, 255, 0]

plt.subplot(121), plt.imshow(cl1, cmap="gray")
plt.subplot(122), plt.imshow(dst, cmap="gray")

cv2.imwrite("subpixel555.png", img)

# TODO elimate noise!!
```
